```python
# sanity_check_same_qid_different_queries.py
import pandas as pd

print("="*100)
print("正确性检查：同一个 Question_ID 必须对应【多个不同 user_query】，但【相同答案】")
print("="*100)

for path in ["train_final_CLEAN.jsonl", "test_final_CLEAN.jsonl"]:
    print(f"\n正在检查 → {path}")
    df = pd.read_json(path, lines=True)
    
    required_cols = ['Question_ID', 'user_query', 'true_a_i']
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        print(f"  缺少必要列: {missing}")
        continue
    
    df['Question_ID'] = df['Question_ID'].astype(str)
    df['user_query'] = df['user_query'].astype(str).str.strip()
    df['true_a_i'] = df['true_a_i'].astype(str).str.strip()
    
    issues = []
    perfect_count = 0
    total_qids = df['Question_ID'].nunique()
    
    for qid, group in df.groupby('Question_ID'):
        queries = group['user_query'].unique()
        answers = group['true_a_i'].unique()
        n_samples = len(group)
        
        # 规则1：必须有多条样本（至少2条才算 paraphrase）
        # 规则2：所有 user_query 必须不同
        # 规则3：所有 true_a_i 必须完全相同
        if n_samples < 2:
            issues.append((qid, "ONLY_ONE_SAMPLE", n_samples, queries[0] if len(queries)>0 else ""))
        elif len(queries) < n_samples:
            issues.append((qid, "DUPLICATE_QUERIES", n_samples, queries.tolist()))
        elif len(answers) > 1:
            issues.append((qid, "DIFFERENT_ANSWERS", n_samples, answers.tolist()))
        else:
            perfect_count += 1  # 完全符合：多条不同query，同一答案
    
    print(f"  总 Question_ID        : {total_qids:,}")
    print(f"  完美符合的题目        : {perfect_count:,}  ({perfect_count/total_qids:.1%})")
    print(f"  有问题的题目          : {len(issues):,}")
    
    if issues:
        print("\n  问题详情（前10个）：")
        for qid, err_type, n, detail in issues[:10]:
            if err_type == "ONLY_ONE_SAMPLE":
                print(f"    [单一样本] Question_ID={qid} 只有 {n} 条，应有多个 paraphrase")
            elif err_type == "DUPLICATE_QUERIES":
                print(f"    [重复问题] Question_ID={qid} 有 {n} 条但只有 {len(detail)} 种不同 query")
            elif err_type == "DIFFERENT_ANSWERS":
                print(f"    [答案不一致] Question_ID={qid} 有多个不同答案: {detail[:3]}...")
        if len(issues) > 10:
            print(f"    ... 还有 {len(issues)-10} 个问题未显示")
    else:
        print("  PERFECT! 所有 Question_ID 都满足：多个不同 user_query + 完全相同的 golden answer")

print("\n" + "="*100)
print("检查完成！")
print("如果看到大量 '完美符合的题目'，说明你的数据结构完全正确，可以直接开训！")
```
